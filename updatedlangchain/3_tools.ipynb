```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
load_dotenv()
```

```python
# Setup OpenRouter as an OpenAI-compatible provider
llm = ChatOpenAI(
    model="openai/gpt-oss-120b",  # Use the OpenRouter model string (e.g., "google/gemini-2.0-flash-001")
    openai_api_key=os.getenv("OPENROUTER_API_KEY"),
    base_url="https://openrouter.ai/api/v1",
    # This enables the "reasoning" feature you liked in your snippet
    extra_body={"reasoning": {"enabled": True}} 
)
```

```python
from langchain.tools import tool

@tool
def get_weather(location: str)->str:
    """Get the weather of location"""
    return f"It's sunny in {location}"


model_with_tools = llm.bind_tools([get_weather])
```

```python
response = model_with_tools.invoke("what is delhi's weather?")
print(response)
```

### Tool Execution loop

```python
messages = [{"role":"user", "content":"What's weather in Boston"}]
ai_msg = model_with_tools.invoke(messages)
messages.append(ai_msg)

# print(ai_msg)

for tool_call in ai_msg.tool_calls:
    tool_result = get_weather.invoke(tool_call)
    messages.append(tool_result)


final_response = model_with_tools.invoke(messages)
print(final_response)
```
