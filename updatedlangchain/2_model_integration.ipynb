```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
load_dotenv()
```

```python
# Setup OpenRouter as an OpenAI-compatible provider
llm = ChatOpenAI(
    model="openai/gpt-oss-120b",  # Use the OpenRouter model string (e.g., "google/gemini-2.0-flash-001")
    openai_api_key=os.getenv("OPENROUTER_API_KEY"),
    base_url="https://openrouter.ai/api/v1",
    # This enables the "reasoning" feature you liked in your snippet
    extra_body={"reasoning": {"enabled": True}} 
)
```

```python
llm
```

```python
response = llm.invoke("Hey buddy")
```

```python
response
```

### Streaming

```python
for chunk in llm.stream("write me 3 lines on india's versetile culture"):
    print(chunk.text, end = "", flush=True)
```

### Batch
In case of muliple input, process can be done parallely

```python
responses = llm.batch([
    "why parrot eat chilies",
    "why smog happens",
    "does money eat meat"
])

for response in responses:
    print(response)
```
